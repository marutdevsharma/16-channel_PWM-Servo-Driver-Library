```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/Puzzle Cube-samples/99.jpg')
plt.imshow(image)
plt.xticks([])
plt.yticks([])
plt.show
```

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
canny_edges = cv2.Canny(grey_image, 30, 200)
M = cv2.moments(canny_edges)
print(M)

cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
print(cx, cy)


circle = cv2.circle(image, (cx, cy), 5, (255, 255, 255), -1)
cv2.putText(circle, 'center', (cx - 25, cy - 25), cv2.FONT_HERSHEY_SIMPLEX,0.5, (255, 255, 255), 2)
plt.imshow(circle)

fig = plt.figure(figsize=(5, 5))


plt.imshow(image)
plt.xticks([])
plt.yticks([])
plt.show()
```
